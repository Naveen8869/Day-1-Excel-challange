5.Password Retry system

```python
correct_password = "python123"
attempt =3

while attempt>0:
  guess = input(f"enter password ({attempt} attempt left): ")

  if guess == correct_password:
    print("Access Granted ")
    break
  else:
    attempt -=1
    print("incorrect password.")
  if attempt ==0:
    print("Access denied. you are locked out.")
```
