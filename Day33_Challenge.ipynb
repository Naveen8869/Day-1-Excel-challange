1 : Number Guessing Game
    

```
# store a secret number(...7)
""" Concept: infinite while + If"""

import random

number = random.randint(0,7)
guess = 0
print('guess a number between 0 and 7')

while guess!=number:
  guess = int(input('enter guess: '))

  if guess<number:
    print('Higher!')
  elif guess>number:
    print('Lower!')
print('you got it!')
```

2 : Reverse a Number

```
# while Loop

n = 1234
reverse = 0
while n>0:
  lase_d = n%10
  reverse = reverse * 10+lase_d
  n = n // 10
print(reverse)
```
