Tasks

```python
Total_sum = salary+rent+food+transport+entertainment
print(Total_sum)
```

Task 2; Savings Calculation

```python
Savings = salary-Total_sum
print(Savings)
```

```python
greater = rent>food
print(greater)
```

```python
less = transport<entertainment
print(less)
```
